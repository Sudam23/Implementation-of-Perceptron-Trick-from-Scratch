```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Dataset Read
dataset= pd.read_csv('skp_annotated_points_np_200.csv')
dataset
```

```python
# Extract X and y
X= dataset[['x','y']].values
X.shape
X
```

```python
y= np.where(dataset['labels']== 3, 1, -1)
y.shape
```

```python
#Scatterplot
plt.scatter(X[:,0],X[:,1], c= ['skyblue' if i == 1 else 'green' for i in y])
plt.title('Data Points')
```

```python
X_new = np.hstack((X,np.ones((200,1))))
```

```python
def perceptron(X_tr,y_tr,w):
     for t in range(1000):
         rd=np.random.permutation(list(range(0,200)))
         for i in rd:
             v= y_tr[i]* np.dot(w,X_tr[i])
             if v<=0: 
                 w += y_tr[i]*X_tr[i] 
     return w           
```

```python
# Randomly initialisation

wgt = np.random.rand(3)
n_wgt = perceptron(X_new,y,wgt)
n_wgt
```

```python
lx=np.linspace(0,30)
ly=(-n_wgt[2]-(n_wgt[0]*lx))/n_wgt[1]
plt.scatter(X[:,0],X[:,1],c=['skyblue' if i == 1 else 'green' for i in y])
plt.plot(lx,ly,'k')
```

```python
wgt = np.zeros(3)
n_wgt = perceptron(X_new,y,wgt)
n_wgt
```

```python
# With zeros initialisation

lx=np.linspace(0,30)
ly=(-n_wgt[2]-(n_wgt[0]*lx))/n_wgt[1]
plt.scatter(X[:,0],X[:,1],c=['skyblue' if i == 1 else 'green' for i in y])
plt.plot(lx,ly,'k')
```

```python
# With median initialisation

wgt = np.array([np.median(X_new)]*3)
n_wgt = perceptron(X_new,y,wgt)
n_wgt
```

```python
lx=np.linspace(0,30)
ly=(-n_wgt[2]-(n_wgt[0]*lx))/n_wgt[1]
plt.scatter(X[:,0],X[:,1],c=['skyblue' if i == 1 else 'green' for i in y])
plt.plot(lx,ly,'k')
```
